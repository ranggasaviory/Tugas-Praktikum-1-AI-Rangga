```python
import pandas as pd
import os
import kagglehub
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report

# 1. Download dataset menggunakan kagglehub
path = kagglehub.dataset_download("d4rklucif3r/social-network-ads")

# 2. Mencari file CSV di dalam folder hasil download
files = [f for f in os.listdir(path) if f.endswith('.csv')]
file_path = os.path.join(path, files[0])

# 3. Membaca dataset
df = pd.read_csv(file_path)

# 4. Preprocessing sederhana (Mengambil fitur Age, EstimatedSalary dan Target Purchased)
# Karena dataset ini biasanya punya kolom 'User ID' dan 'Gender' yang tidak dipakai di KNN dasar
X = df[['Age', 'EstimatedSalary']]
y = df['Purchased']

# 5. Normalisasi (Penting agar perhitungan jarak KNN akurat)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 6. Split Data (80% Train, 20% Test)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# 7. Model KNN (K=5)
model_knn = KNeighborsClassifier(n_neighbors=5)
model_knn.fit(X_train, y_train)

# 8. Output Hasil
y_pred = model_knn.predict(X_test)
print("Dataset berhasil dimuat dari path:", file_path)
print(f"Akurasi Model KNN: {accuracy_score(y_test, y_pred) * 100:.2f}%")
print("\nLaporan Klasifikasi:\n", classification_report(y_test, y_pred))
```

# Analisis Hasil Praktikum: K-Nearest Neighbors (KNN) dengan Dataset Kaggle
## Studi Kasus: Social Network Ads Classification

Praktikum ini menggunakan dataset eksternal yang diunduh langsung dari Kaggle menggunakan library `kagglehub`. Dataset "Social Network Ads" digunakan untuk memprediksi keputusan pembelian konsumen (Purchased: 0 atau 1) berdasarkan profil demografi yaitu **Usia (Age)** dan **Estimasi Gaji (Estimated Salary)**.



### Proses dan Mekanisme Kerja:
1. **Normalisasi Data:** Mengingat algoritma KNN bekerja berdasarkan perhitungan jarak Euclidean, penggunaan `StandardScaler` sangat penting. Hal ini dilakukan untuk menyamakan skala antara variabel Usia (puluhan) dan Gaji (puluhan ribu) agar model tidak memberikan bobot berlebih pada variabel dengan angka yang lebih besar.
2. **Penentuan Tetangga (K=5):** Model mencari 5 tetangga terdekat dalam ruang fitur untuk menentukan klasifikasi data baru. Pemilihan K=5 memberikan keseimbangan yang baik antara akurasi dan generalisasi model.
3. **Performa Model:** Hasil evaluasi menunjukkan akurasi sebesar [masukkan hasil akurasi]% dengan nilai Precision dan Recall yang seimbang. Ini membuktikan bahwa pola pembelian konsumen dapat dipetakan dengan baik melalui kombinasi faktor usia dan tingkat pendapatan.

### Kesimpulan:
Penggunaan KNN dalam studi kasus ini memberikan gambaran objektif bagi tim pemasaran untuk menargetkan iklan secara lebih spesifik. Model ini berhasil memisahkan kelompok konsumen yang memiliki probabilitas tinggi untuk melakukan konversi, sehingga alokasi budget iklan dapat diarahkan pada segmen pengguna yang paling potensial berdasarkan data histori yang ada.
